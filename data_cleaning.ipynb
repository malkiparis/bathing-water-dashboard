```python
df = pd.read_excel("/content/drive/MyDrive/Data_Science/bw_assessment_eea_datahub_1990_2024.xlsx",
                   sheet_name=1)  #sheet name_= 1 means accessing the BWD_data_1990_2024
```

```python
# Displays first 5 rows to quickly inspect the dataset structure
df.head()
df.columns
```

```python
# Provides summary: data types, non-null counts, helps identify missing values
df.info()
df.columns
```

```python
# Cleans column names by removing spaces and converting to lowercase for consistency
df.columns = df.columns.str.strip().str.lower()
```

```python
# Removes rows where 'quality' is missing
df = df.dropna(subset=['quality'])
```

```python
# Extracts the actual label
df['qualitylabel'] = df['quality'].str.split(' - ').str[1]
```

```python
# Extracts the 4-digit year from 'season' column using regex and converts to integer
df['year'] = df['season'].astype(str).str.extract(r'(\d{4})').astype(int)
```

```python
# Simplifies long category names into cleaner, readable labels
df['watertype'] = df['bathingwatertype'].replace({
    'coastalBathingWater': 'Coastal',
    'inlandBathingWater': 'Inland'
})
```

```python
#removing extra space from country codes
df['countrycode'] = df['countrycode'].str.strip()
#clean beach names
df['bathingwatername'] = df['bathingwatername'].str.strip()
```

```python
#remove rows without coordinates
df = df.dropna(subset=['lat', 'lon'])
```

```python
# Keeps only relevant columns needed for analysis and dashboard
df = df[['countrycode',
         'bathingwatername',
         'watertype',
         'lat',
         'lon',
         'year',
         'qualitylabel']]
```

```python
# Removes unclear categories
df = df[~df['qualitylabel'].isin(['Not classified', 'Good or Sufficient'])]
```

```python
print(df.shape)  #displays number of rows and columns after cleaning
print(df.head())
print(df['qualitylabel'].value_counts())
```
